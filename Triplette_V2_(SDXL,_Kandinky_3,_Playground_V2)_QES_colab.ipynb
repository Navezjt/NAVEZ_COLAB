```python
#@title Installation
!pip install -q git+https://github.com/huggingface/diffusers.git
!pip install --upgrade transformers accelerate gradio omegaconf invisible-watermark>=0.2.0
!pip install sentencepiece

import os
import random
import uuid
from datetime import datetime
import time

import gradio as gr
import numpy as np
from PIL import Image
import torch
from diffusers import DiffusionPipeline, AutoPipelineForText2Image, StableDiffusionXLPipeline, StableDiffusionXLImg2ImgPipeline

folder_path = "/content/drive/My Drive/Triplette"

# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Load Kandinsky 3 model
kandinsky_3_model = AutoPipelineForText2Image.from_pretrained("kandinsky-community/kandinsky-3", variant="fp16", torch_dtype=torch.float16)
kandinsky_3_model.enable_model_cpu_offload()

# Load Stable Diffusion SDXL model
sd_sdxl_model = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, variant="fp16", use_safetensors=True, add_watermarker=False
).to("cuda")
```

```python
#@title Image Generation Settings
DESCRIPTION = """# Quick-Eyed Sky Triplette: Stable Diffusion SDXL, Kandinsky 3, Playground V2"""
if not torch.cuda.is_available():
    DESCRIPTION += "\n<p>Running on CPU ðŸ¥¶ This demo may not work on CPU.</p>"

MAX_SEED = np.iinfo(np.int32).max
CACHE_EXAMPLES = torch.cuda.is_available() and os.getenv("CACHE_EXAMPLES", "1") == "1"
MAX_IMAGE_SIZE = int(os.getenv("MAX_IMAGE_SIZE", "2048"))
USE_TORCH_COMPILE = os.getenv("USE_TORCH_COMPILE", "0") == "1"
ENABLE_CPU_OFFLOAD = os.getenv("ENABLE_CPU_OFFLOAD", "0") == "1"

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

if torch.cuda.is_available():
    pipe = DiffusionPipeline.from_pretrained(
        "playgroundai/playground-v2-1024px-aesthetic",
        torch_dtype=torch.float16,
        use_safetensors=True,
        add_watermarker=False,
        variant="fp16"
    )
    if ENABLE_CPU_OFFLOAD:
        pipe.enable_model_cpu_offload()
    else:
        pipe.to(device)
        print("Loaded on Device!")

    if USE_TORCH_COMPILE:
        pipe.unet = torch.compile(pipe.unet, mode="reduce-overhead", fullgraph=True)
        print("Model Compiled!")

# New part: Model-specific settings for steps and guidance scale
MODEL_SETTINGS = {
    "PLAY": {"steps": 25, "guidance_scale": 3.0},
    "KAND": {"steps": 50, "guidance_scale": 5.0},
    "SDXL": {"steps": 75, "guidance_scale": 8.0}
}


def save_image(img, model_code, folder_path):
    unique_name = str(uuid.uuid4()) + f"_{model_code}.png"
    full_path = os.path.join(folder_path, unique_name)
    img.save(full_path)
    return full_path

def process_dynamic_prompt(prompt):
    while '{' in prompt:
        start = prompt.find('{')
        end = prompt.find('}', start)
        if end == -1:
            break
        options = prompt[start+1:end].split('|')
        choice = random.choice(options)
        prompt = prompt[:start] + choice + prompt[end+1:]
    return prompt

def contains_dynamic_elements(prompt):
    return '{' in prompt and '}' in prompt

def randomize_seed_fn(seed: int, randomize_seed: bool) -> int:
    if randomize_seed:
        seed = random.randint(0, np.iinfo(np.int32).max)
    return seed

# Modify the save_individual_settings function to accept a dictionary of settings
def save_individual_settings(folder_path, image_num, individual_prompt, original_dynamic_prompt, negative_prompt, seed, width, height, settings, randomize_seed, session_id):
    filename = f"settings_{session_id}_image_{image_num}.txt"
    full_path = os.path.join(folder_path, filename)
    with open(full_path, 'w') as file:
        file.write(f"Original Dynamic Prompt: {original_dynamic_prompt}\n")
        file.write(f"Processed Prompt: {individual_prompt}\n")
        file.write(f"Negative Prompt: {negative_prompt if negative_prompt else 'None'}\n")
        file.write(f"Seed: {seed}\n")
        file.write(f"Width: {width}\n")
        file.write(f"Height: {height}\n")
        file.write(f"Guidance and Steps Settings:\n")
        for model_code, model_settings in settings.items():
            file.write(f"  Model {model_code}:\n")
            file.write(f"    Guidance Scale: {model_settings['guidance_scale']}\n")
            file.write(f"    Steps: {model_settings['steps']}\n")
        file.write(f"Randomize Seed: {'Yes' if randomize_seed else 'No'}\n")
    print(f"Settings for image {image_num} saved to {full_path}")
    return individual_prompt

# Update the generate function to pass guidance and steps settings for all models
def generate(
    prompt: str,
    negative_prompt: str = "",
    use_negative_prompt: bool = False,
    seed: int = 0,
    width: int = 1280,
    height: int = 1024,
    guidance_scale_play: float = 7.0,
    guidance_scale_kand: float = 5.0,
    guidance_scale_sdxl: float = 8.0,
    randomize_seed: bool = False,
    num_images: int = 1,
    use_playground_v2: bool = True,
    use_kandinsky_3: bool = True,
    use_sd_sdxl: bool = True,
    steps_play: int = 25,
    steps_kand: int = 50,
    steps_sdxl: int = 75
):
    folder_path = "/content/drive/My Drive/Triplette"
    output_data = []
    session_id = int(time.time())  # Unique identifier for each session

    for i in range(num_images):
        seed = int(randomize_seed_fn(seed, randomize_seed))
        generator = torch.Generator().manual_seed(seed)

        if not use_negative_prompt:
            negative_prompt = None

        original_dynamic_prompt = prompt
        processed_prompt = process_dynamic_prompt(prompt) if contains_dynamic_elements(prompt) else prompt

        model_pipelines = {
            "PLAY": pipe,
            "KAND": kandinsky_3_model,
            "SDXL": sd_sdxl_model
        }

        # Dictionary to store guidance and steps settings for each model
        model_settings = {
            "PLAY": {"guidance_scale": guidance_scale_play, "steps": steps_play},
            "KAND": {"guidance_scale": guidance_scale_kand, "steps": steps_kand},
            "SDXL": {"guidance_scale": guidance_scale_sdxl, "steps": steps_sdxl}
        }

        for model_code, pipeline in model_pipelines.items():
            if (model_code == "PLAY" and use_playground_v2) or \
               (model_code == "KAND" and use_kandinsky_3) or \
               (model_code == "SDXL" and use_sd_sdxl):

                print(f"Generating image using {model_code} model...")

                model_steps = model_settings[model_code]["steps"]
                model_guidance_scale = model_settings[model_code]["guidance_scale"]

                images = pipeline(
                    prompt=processed_prompt,
                    width=width,
                    height=height,
                    guidance_scale=model_guidance_scale,
                    num_inference_steps=model_steps,
                    generator=generator,
                ).images

                for img in images:
                    image_path = save_image(img, model_code, folder_path)
                    output_data.append(image_path)

        # Pass the model_settings dictionary to save_individual_settings
        save_individual_settings(folder_path, i + 1, processed_prompt, original_dynamic_prompt, negative_prompt, seed, width, height, model_settings, randomize_seed, session_id)

    print("Image generation completed.")
    return output_data, seed
```

```python
#@title Gradio Settings (when ready, click on the link "Running on Public URL")
css = '''
.gradio-container {
    max-width: 1680px !important;
}
h1 {
    text-align: center;
}
'''

with gr.Blocks(css=css) as demo:
    gr.Markdown(DESCRIPTION)

    # Model selection checkboxes
    with gr.Accordion("Model Selection", open=True):
        with gr.Row():
            use_playground_v2 = gr.Checkbox(label="Use Playground V2", value=True)
            use_kandinsky_3 = gr.Checkbox(label="Use Kandinsky 3", value=True)
            use_sd_sdxl = gr.Checkbox(label="Use Stable Diffusion SDXL", value=True)

    with gr.Group():
        with gr.Row():
            prompt = gr.Text(
                label="Prompt",
                show_label=False,
                placeholder="Enter your prompt",
                container=False,
            )
            run_button = gr.Button("Run", scale=0)
        result = gr.Gallery(label="Result", show_label=False)

    with gr.Accordion("Advanced options", open=True):
        with gr.Row():
            use_negative_prompt = gr.Checkbox(label="Use negative prompt", value=True)
            negative_prompt = gr.Text(
                label="Negative prompt",
                max_lines=1,
                placeholder="Enter a negative prompt",
                visible=True,
            )
        seed = gr.Slider(
            label="Seed",
            minimum=0,
            maximum=MAX_SEED,
            step=1,
            value=0,
        )
        randomize_seed = gr.Checkbox(label="Randomize seed", value=True)
        with gr.Row(visible=True):
            width = gr.Slider(
                label="Width",
                minimum=256,
                maximum=MAX_IMAGE_SIZE,
                step=32,
                value=1024,
            )
            height = gr.Slider(
                label="Height",
                minimum=256,
                maximum=MAX_IMAGE_SIZE,
                step=32,
                value=1024,
            )

    # Steps and Guidance Scale sliders under model selection for each model
    with gr.Group():
        gr.Markdown("<h3>PLAY</h3>")
        guidance_scale_play = gr.Slider(
            label="Guidance Scale",
            minimum=0.1,
            maximum=20,
            step=0.1,
            value=3.0,
        )
        steps_play = gr.Slider(minimum=1, maximum=200, step=1, label="Steps", value=25)

    with gr.Group():
        gr.Markdown("<h3>KAND</h3>")
        guidance_scale_kand = gr.Slider(
            label="Guidance Scale",
            minimum=0.1,
            maximum=20,
            step=0.1,
            value=5.0,
        )
        steps_kand = gr.Slider(minimum=1, maximum=200, step=1, label="Steps", value=30)

    with gr.Group():
        gr.Markdown("<h3>SDXL</h3>")
        guidance_scale_sdxl = gr.Slider(
            label="Guidance Scale",
            minimum=0.1,
            maximum=20,
            step=0.1,
            value=8.0,
        )
        steps_sdxl = gr.Slider(minimum=1, maximum=200, step=1, label="Steps", value=50)

    num_images = gr.Slider(minimum=1, maximum=200, step=1, label="Number of Images", value=6)

    use_negative_prompt.change(
        fn=lambda x: gr.update(visible=x),
        inputs=use_negative_prompt,
        outputs=negative_prompt,
        api_name=False,
    )

    gr.on(
        triggers=[
            prompt.submit,
            negative_prompt.submit,
            run_button.click,
        ],
        fn=generate,
        inputs=[
            prompt,
            negative_prompt,
            use_negative_prompt,
            seed,
            width,
            height,
            guidance_scale_play,  # Use separate guidance scales for each model
            guidance_scale_kand,
            guidance_scale_sdxl,
            randomize_seed,
            num_images,
            use_playground_v2,
            use_kandinsky_3,
            use_sd_sdxl,
            steps_play,  # Use separate steps for each model
            steps_kand,
            steps_sdxl,
        ],
        outputs=[result, seed],
        api_name="run",
    )

if __name__ == "__main__":
    demo.launch(share=True, debug=True, inline=False)

```
